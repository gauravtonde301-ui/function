# List Sum: Write a function that takes a list of numbers and prints their total sum.

```python
def list_sum(num):
    total=0
    for i in num:
        total=total+i
    print(total)    
```

```python
list_sum([1,2,3,4,5])
```

# Vowel Counter: Create a function count_vowels(word) that loops through a string and prints the number of vowels found.

```python
def count_vowel(word):
    count=0
    for j in word:
        if j=='a' or j=='e' or j=='i' or j=='o' or j=='u':
            count+=1
    print(count)
```

```python
count_vowel('gaurav')
```

# Password Validator: Write a function that checks if a password is longer than 8 characters AND contains the symbol "@".

# Nested Logic: Write a function that takes a list of ages and uses the even_odd logic from your notebook to count how many even-numbered ages exist.


```python
def ages(number):
    even_number=0
    for i in number:
        if i%2==0:
            even_number+=1
    print('the even ages are :', even_number)
            
```

```python
ages([1,2,3,4,5,6,7,8,9])
```
